```python
import cv2 as cv
import matplotlib.pyplot as plt
cl = cv.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
def detect_faces_cascade(img, color = (0, 0, 255)):
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

    faces = cl.detectMultiScale(gray, 1.3, 5)

    
    #for (x,y,w,h) in faces:
    #    cv.rectangle(img, (x, y), (x+w, y+h), color, 2)   

    return faces, img

def crop_face(img, faces):
    # Assuming that exists only one face in the image

    (x, y, w, h) = faces[0]

    croped_img = img[y:y+w, x:x+h]

    return croped_img
```

```python
img = cv.imread("data/angry/0.bmp")
```

```python
faces, img = detect_faces_cascade(img)
```

```python
plt.imshow(crop_face(img, faces))
```
